### Extracting Frames form videos

```python
BASE = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"
datasets = {"FF++": os.path.join(BASE, "FF++"),
            "DFD": os.path.join(BASE, "DFD")}
labels = ["0_real", "1_fake"]

FRAMES_PER_VIDEO = 60  # Set number of frames per video
VERBOSE = True

def extract_fixed_frames(video_path, out_dir, frames_target=FRAMES_PER_VIDEO):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return 0, "cannot_open"

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames == 0:
        return 0, "no_frames"

    step = max(1, total_frames // frames_target)
    os.makedirs(out_dir, exist_ok=True)

    saved = 0
    frame_idx = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % step == 0:
            fname = f"frame_{saved:04d}.jpg"
            out_path = os.path.join(out_dir, fname)
            cv2.imwrite(out_path, frame)
            saved += 1
            if saved >= frames_target:
                break
        frame_idx += 1

    cap.release()
    return saved, None

summary = []

for ds_name, ds_path in datasets.items():
    print(f"\nProcessing dataset: {ds_name}")
    for label in labels:
        video_root = os.path.join(ds_path, label)
        frames_root = os.path.join(ds_path, f"{label}_frames_all")
        os.makedirs(frames_root, exist_ok=True)

        if not os.path.isdir(video_root):
            continue

        videos = [f for f in os.listdir(video_root) if f.lower().endswith(('.mp4', '.avi', '.mov', '.mkv'))]
        print(f"  {label}: {len(videos)} videos")

        for v in tqdm(videos, desc=f"{ds_name}/{label}", leave=False):
            vpath = os.path.join(video_root, v)
            vname = os.path.splitext(v)[0]
            out_dir = os.path.join(frames_root, vname)
            saved, error = extract_fixed_frames(vpath, out_dir, FRAMES_PER_VIDEO)
            summary.append((ds_name, label, v, saved, error))
```

```python
# === Paths & setup ===
BASE = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"

datasets = {
    "FF++": os.path.join(BASE, "FF++"),
    "DFD": os.path.join(BASE, "DFD")
}

labels = ["0_real_frames_all", "1_fake_frames_all"]  # your extracted frame folders
OUT_BASE = os.path.join(BASE, "faces_cropped")
FAILED_BASE = os.path.join(BASE, "faces_failed")
os.makedirs(OUT_BASE, exist_ok=True)
os.makedirs(FAILED_BASE, exist_ok=True)

detector = MTCNN()

# === Face detection & cropping function ===
def detect_and_crop_faces(src_folder, out_folder, fail_folder):
    cropped_count = 0
    failed_count = 0

    for root, _, files in os.walk(src_folder):
        img_files = [f for f in files if f.lower().endswith(('.jpg', '.png'))]
        if not img_files:
            continue

        rel_path = os.path.relpath(root, src_folder)
        save_dir = os.path.join(out_folder, rel_path)
        fail_dir = os.path.join(fail_folder, rel_path)
        os.makedirs(save_dir, exist_ok=True)
        os.makedirs(fail_dir, exist_ok=True)

        for fname in tqdm(img_files, desc=f"Processing {os.path.basename(root)}", leave=False):
            src_path = os.path.join(root, fname)
            out_path = os.path.join(save_dir, fname)
            fail_path = os.path.join(fail_dir, fname)

            img = cv2.imread(src_path)
            if img is None:
                continue

            faces = detector.detect_faces(img)
            if len(faces) == 0:
                failed_count += 1
                cv2.imwrite(fail_path, img)
                continue

            x, y, w, h = faces[0]['box']
            x, y = max(0, x), max(0, y)
            face = img[y:y+h, x:x+w]
            face = cv2.resize(face, (224, 224))
            cv2.imwrite(out_path, face)
            cropped_count += 1

    return cropped_count, failed_count

# === Run on all datasets ===
for ds_name, ds_path in datasets.items():
    for label in labels:
        src_dir = os.path.join(ds_path, label)
        out_dir = os.path.join(OUT_BASE, f"{ds_name}_{label}_faces")
        fail_dir = os.path.join(FAILED_BASE, f"{ds_name}_{label}_failed")
        print(f"\n🔍 Processing frames from: {src_dir}")
        cropped, failed = detect_and_crop_faces(src_dir, out_dir, fail_dir)
        print(f"✅ {ds_name} {label}: Cropped={cropped}, Failed={failed}")

print("\n✅ Face detection & cropping complete!")
print(f"Cropped faces saved under: {OUT_BASE}")
print(f"Failed detections saved under: {FAILED_BASE}")
```

### Preprocessing and Normalization

```python
# === Paths ===
BASE = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"
faces_folder = os.path.join(BASE, "faces_cropped")
PREPROCESSED_FOLDER = os.path.join(BASE, "preprocessed_faces")
os.makedirs(PREPROCESSED_FOLDER, exist_ok=True)

IMG_SIZE = (224, 224) 

# === Recursive Preprocessing ===
def preprocess_and_normalize(src_folder, out_folder, img_size=IMG_SIZE):
    total_images = 0

    for ds_label in os.listdir(src_folder):
        ds_path = os.path.join(src_folder, ds_label)
        if not os.path.isdir(ds_path):
            continue

        # Walk through ALL subfolders
        for root, _, files in os.walk(ds_path):
            img_files = [f for f in files if f.lower().endswith(('.jpg', '.png'))]
            if not img_files:
                continue

            # Mirror the same folder structure in output
            rel_path = os.path.relpath(root, src_folder)
            save_dir = os.path.join(out_folder, rel_path)
            os.makedirs(save_dir, exist_ok=True)

            # Process and save each image
            for fname in tqdm(img_files, desc=f"Processing {rel_path}", leave=False):
                src_path = os.path.join(root, fname)
                img = cv2.imread(src_path)
                if img is None:
                    continue

                img = cv2.resize(img, img_size)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = img.astype(np.float32) / 255.0

                save_path = os.path.join(save_dir, fname.replace('.jpg', '.npy').replace('.png', '.npy'))
                np.save(save_path, img)
                total_images += 1

    return total_images

# === Run Preprocessing ===
total = preprocess_and_normalize(faces_folder, PREPROCESSED_FOLDER)
print(f"\n✅ Preprocessing and normalization complete! Total images processed: {total}")
print(f"Preprocessed faces saved at: {PREPROCESSED_FOLDER}")
```

### Deep feature extraction

```python
import os
import numpy as np
from tqdm import tqdm
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.applications.densenet import preprocess_input

# === PATH CONFIG ===
BASE_DIR = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"
INPUT_DIR = os.path.join(BASE_DIR, "preprocessed_faces")   # your .npy faces
OUTPUT_DIR = os.path.join(BASE_DIR, "deep_features", "DenseNet121")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# === LOAD MODEL ===
print("Loading DenseNet121 model (without top)...")
model = DenseNet121(weights='imagenet', include_top=False, pooling='avg')
print("✅ Model loaded successfully!")

def extract_features_from_folder(src_dir, dst_dir):
    total_vectors = 0
    folder_name = os.path.basename(src_dir)
    print(f"\n🚀 Starting extraction for: {folder_name}")

    for root, _, files in os.walk(src_dir):
        npy_files = [f for f in files if f.lower().endswith('.npy')]
        if not npy_files:
            continue

        rel_path = os.path.relpath(root, src_dir)
        save_dir = os.path.join(dst_dir, rel_path)
        os.makedirs(save_dir, exist_ok=True)

        all_features = []

        for fname in tqdm(npy_files, desc=f"Extracting from {os.path.basename(root)}", leave=False):
            fpath = os.path.join(root, fname)
            try:
                img = np.load(fpath)           # shape (224,224,3), normalized [0,1]
                
                if img.ndim == 3:
                    img = np.expand_dims(img, axis=0)   # (1,224,224,3)
                elif img.ndim != 4:
                    print(f"⚠️ Skipped {fname}, unexpected shape {img.shape}")
                    continue

                # DenseNet preprocess expects 0–255 BGR-ish, but keras handles channels
                img_pp = preprocess_input(img * 255.0)
                features = model.predict(img_pp, verbose=0)  # (1, 1024) for DenseNet121
                all_features.extend(features)
                total_vectors += features.shape[0]

            except Exception as e:
                print(f"❌ Error processing {fname}: {e}")
                continue

        if all_features:
            feature_array = np.array(all_features)
            out_path = os.path.join(save_dir, "features.npy")
            np.save(out_path, feature_array)
            print(f"✅ Saved features for {os.path.basename(root)}: {feature_array.shape}")

    print(f"🎯 Completed extraction for {folder_name} | Total feature vectors: {total_vectors}\n")
    return total_vectors

# === RUN EXTRACTION FOR YOUR MAIN FOLDERS ===
print("\n=== Starting DenseNet121 Deep Feature Extraction ===")

main_folders = [
    "DFD_0_real_frames_all_faces",
    "DFD_1_fake_frames_all_faces",
    "FF++_0_real_frames_all_faces",
    "FF++_1_fake_frames_all_faces"
]

grand_total = 0
for folder in main_folders:
    src_path = os.path.join(INPUT_DIR, folder)
    dst_path = os.path.join(OUTPUT_DIR, folder)
    count = extract_features_from_folder(src_path, dst_path)
    grand_total += count

print("\n✅ DenseNet121 deep feature extraction complete for ALL datasets!")
print(f"📊 Grand total feature vectors extracted: {grand_total}")
print(f"💾 Features saved under: {OUTPUT_DIR}")
```

### Handcrafted Feature 

```python
# === PATH CONFIG ===
BASE_DIR = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"
INPUT_DIR = os.path.join(BASE_DIR, "preprocessed_faces")
OUTPUT_DIR = os.path.join(BASE_DIR, "handcrafted_features")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# === HYPERPARAMETERS FOR FEATURES ===
LBP_POINTS = 8
LBP_RADIUS = 1
HOG_PIXELS_PER_CELL = (8, 8)
HOG_CELLS_PER_BLOCK = (2, 2)

# === FEATURE EXTRACTION FROM ONE IMAGE ===
def extract_handcrafted_features(img):
    # Convert to grayscale
    gray = rgb2gray(img)

    #  LBP (Local Binary Pattern)
    lbp = local_binary_pattern(gray, LBP_POINTS, LBP_RADIUS, method="uniform")
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, LBP_POINTS + 3), density=True)

    # HOG (Histogram of Oriented Gradients)
    hog_features = hog(
        gray,
        pixels_per_cell=HOG_PIXELS_PER_CELL,
        cells_per_block=HOG_CELLS_PER_BLOCK,
        feature_vector=True,
    )

    # Color moments (mean, std, skewness, kurtosis per channel)
    color_feats = []
    for i in range(img.shape[2]):
        channel = img[:, :, i].ravel()
        color_feats.extend([np.mean(channel), np.std(channel), skew(channel), kurtosis(channel)])

    # Combine all
    features = np.concatenate([lbp_hist, hog_features, color_feats])
    return features


# === RECURSIVE FEATURE EXTRACTION FUNCTION ===
def extract_handcrafted_recursive(src_dir, dst_dir):
    for root, _, files in os.walk(src_dir):
        npy_files = [f for f in files if f.lower().endswith(".npy")]
        if not npy_files:
            continue

        rel_path = os.path.relpath(root, src_dir)
        save_dir = os.path.join(dst_dir, rel_path)
        os.makedirs(save_dir, exist_ok=True)

        all_features = []

        for fname in tqdm(npy_files, desc=f"Extracting from {os.path.basename(root)}", leave=False):
            fpath = os.path.join(root, fname)
            try:
                frame = np.load(fpath)
                if frame.ndim == 3:  # (224, 224, 3)
                    features = extract_handcrafted_features(frame)
                    all_features.append(features)
                else:
                    print(f"⚠️ Unexpected shape {frame.shape} in {fname}")
            except Exception as e:
                print(f"Error processing {fname}: {e}")

        if all_features:
            feature_array = np.array(all_features)
            out_path = os.path.join(save_dir, "handcrafted_features.npy")
            np.save(out_path, feature_array)
            print(f"✅ Saved handcrafted features for {os.path.basename(root)}: {feature_array.shape}")


# === RUN EXTRACTION ===
print("\nStarting handcrafted feature extraction...")
extract_handcrafted_recursive(INPUT_DIR, OUTPUT_DIR)
print("\n✅ Handcrafted feature extraction complete!")
print(f"Features saved under: {OUTPUT_DIR}")
```

### Fuse deepfeatures and handcrafted features

```python
import os
import numpy as np
from tqdm import tqdm

# === PATH CONFIG ===
BASE_DIR = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets"
DEEP_DIR = os.path.join(BASE_DIR, "deep_features", "DenseNet121")
HANDCRAFT_DIR = os.path.join(BASE_DIR, "handcrafted_features")
OUTPUT_DIR = os.path.join(BASE_DIR, "fused_features")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# === RECURSIVE FUSION ===
def fuse_features_recursive(deep_root, hand_root, out_root, method="concat", alpha=0.5):
    for root, _, files in os.walk(deep_root):
        if "features.npy" not in files:
            continue

        rel_path = os.path.relpath(root, deep_root)
        deep_path = os.path.join(root, "features.npy")
        hand_path = os.path.join(hand_root, rel_path, "handcrafted_features.npy")

        if not os.path.exists(hand_path):
            print(f"⚠️ Missing handcrafted features for {rel_path}")
            continue

        deep_feat = np.load(deep_path)
        hand_feat = np.load(hand_path)

        n = min(deep_feat.shape[0], hand_feat.shape[0])
        deep_feat, hand_feat = deep_feat[:n], hand_feat[:n]

        if method == "concat":
            fused = np.concatenate([deep_feat, hand_feat], axis=1)
        elif method == "average":
            fused = (deep_feat + hand_feat) / 2
        elif method == "weighted":
            fused = alpha * deep_feat + (1 - alpha) * hand_feat
        else:
            raise ValueError("Invalid method.")

        save_dir = os.path.join(out_root, rel_path)
        os.makedirs(save_dir, exist_ok=True)
        np.save(os.path.join(save_dir, "fused_features.npy"), fused)

        print(f"✅ Fused: {rel_path} | Shape = {fused.shape}")

# === RUN ===
print("🔄 Starting recursive feature fusion...")
fuse_features_recursive(DEEP_DIR, HANDCRAFT_DIR, OUTPUT_DIR, method="concat")
print("\n🎯 Recursive fusion complete! Saved in:", OUTPUT_DIR)
```

```python
# === Path configuration ===
BASE_DIR = r"C:\Users\Mahima\OneDrive\Desktop\NewDataset(FF++andDFD)3\datasets\fused_features"

all_features = []
all_labels = []

# Map labels based on folder names
label_map = {
    "0_real_frames_all_faces": 0,
    "1_fake_frames_all_faces": 1
}

# === Recursively collect all fused feature files ===
for dataset_name in os.listdir(BASE_DIR):
    ds_path = os.path.join(BASE_DIR, dataset_name)
    if not os.path.isdir(ds_path):
        continue
    
    # Determine label (0 or 1)
    label = None
    for key in label_map:
        if key in dataset_name:
            label = label_map[key]
            break
    if label is None:
        continue

    for root, _, files in os.walk(ds_path):
        for f in files:
            if f == "fused_features.npy":
                fpath = os.path.join(root, f)
                try:
                    data = np.load(fpath)
                    if len(data.shape) == 1:
                        data = data.reshape(1, -1)
                    all_features.append(data)
                    all_labels.extend([label] * len(data))
                except Exception as e:
                    print(f"⚠️ Error loading {fpath}: {e}")

# === Combine all features and labels ===
X = np.vstack(all_features)
y = np.array(all_labels)

print(f"\n✅ Combined fused features shape: {X.shape}")
print(f"✅ Combined labels shape: {y.shape}")

# === Save combined arrays for training ===
np.save(os.path.join(BASE_DIR, "fused_features_combined.npy"), X)
np.save(os.path.join(BASE_DIR, "fused_labels_combined.npy"), y)
print(f"\n💾 Saved combined fused features and labels under {BASE_DIR}")
```
